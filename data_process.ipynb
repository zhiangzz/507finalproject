```python

Recommend= ["imdb",["genre",None,None,None],
              ["runtime",None,None,None],
              ["release_year",None,None,None],
      "golden globe award",["award category",None,None,None],
                      ["award year",None,None,None],
           ]
Genre=('Drama','Crime','Action', 'Adventure','Biography','History','Romance','Western','Thriller','Family','Animation','Comedy',' Mystery',
 'Music','War','Horror')

imdb_runtime=("<=60","60-120","120-180",">180")
release_year=("2010-2020","2000-2010","1990-2000","1970-1990","before 1970")
award_category=("movie","television")
movie=('Best Performance by an Actress in a Supporting Role in any Motion Picture', 'Best Performance by an Actor in a Supporting Role in any Motion Picture', 'Best Director - Motion Picture', 'Best Screenplay - Motion Picture', 'Best Motion Picture - Foreign Language', 'Best Performance by an Actress in a Motion Picture - Drama', 'Best Performance by an Actor in a Motion Picture - Drama', 'Best Performance by an Actor in a Motion Picture - Musical or Comedy', 'Best Performance by an Actress in a Motion Picture - Musical or Comedy')
television=('Television Producer/Director', 'Actor In A Television Series', 'Actress In A Television Series', 'Actor In A Television Series - Musical Or Comedy', 'Actor In A Television Series - Drama', 'Actress In A Television Series - Drama', 'Actress In A Television Series - Musical Or Comedy', 'Best Performance by an Actress In A Television Series - Drama', 'Best Performance by an Actor In A Television Series - Drama', 'Best Performance by an Actress in a Television Series - Musical or Comedy', 'Best Performance by an Actor in a Television Series - Musical or Comedy', 'Actor In A Supporting Role - Television Series', 'Actress In A Supporting Role - Television Series', 'Actress In A Supporting Role - Series Or Television Movie', 'Actress In A Leading Role - Drama Series Or Television Movie', 'Actor In A Leading Role - Drama Series Or Television Movie', 'Actor In A Leading Role - Musical Or Comedy Series Or Television Movie', 'Actor In A Supporting Role - Series Or Television Movie', 'Actress In A Leading Role - Musical Or Comedy Series Or Television Movie', 'Best Performance by an Actress in a Supporting Role in a Series, Limited Series or Motion Picture Made for Television', 'Best Performance by an Actor in a Supporting Role in a Series, Limited Series or Motion Picture Made for Television', 'Best Television Series - Drama', 'Best Performance by an Actress in a Limited Series or a Motion Picture Made for Television', 'Best Performance by an Actor in a Limited Series or a Motion Picture Made for Television')
```

```python
#data pre-processing:
def Release(data):
    release={
        "2010-2020":[],
        "2000-2010":[],
        "1990-2000":[],
        "1970-1990":[],
        "before 1970":[],
        "other":[]
    }
    for i in range(len(data["Released_Year"])):
        try:
            if 2010<int(data["Released_Year"][i])<=2020:
                release["2010-2020"].append(data["Series_Title"][i])
            elif 2000<int(data["Released_Year"][i])<=2010:
                release["2000-2010"].append(data["Series_Title"][i])
            elif 1990<int(data["Released_Year"][i])<=2000:
                release["1990-2000"].append(data["Series_Title"][i])
            elif 1970<int(data["Released_Year"][i])<=1990:
                release["1970-1990"].append(data["Series_Title"][i])
            elif int(data["Released_Year"][i])<=1970:
                release["before 1970"].append(data["Series_Title"][i])
        except:
            release["other"].append(data["Series_Title"][i])
    return release
```

```python
leaf_release=Release(imdb)
```

```python
def Runtime(data):
    runtime={
        "<=60":[],
        "60-120":[],
        "120-180":[],
        ">180":[]
    }
    for i in range(len(data["Runtime"])):
        if int(data["Runtime"][i])<=60:
            runtime["<=60"].append(data["Series_Title"][i])
        elif 60<int(data["Runtime"][i])<=120:
            runtime["60-120"].append(data["Series_Title"][i])
        elif 120<int(data["Runtime"][i])<=180:
            runtime["120-180"].append(data["Series_Title"][i])
        elif int(data["Runtime"][i])>180:
            runtime[">180"].append(data["Series_Title"][i])
    return runtime
        
            
    
```

```python
def category(data):
    category={"movie":[],
              "television":[]}
    for i in range(len(data["category"])):
        if data["category"][i] in movie:
            if data["film"][i] not in category["movie"]:
                category["movie"].append(data["film"][i])
        elif data["category"][i] in television:
            category["television"].append(data["film"][i])
    return category
            
                            
```

```python
leaf_category=category(golden_data)
```

```python
leaf_runtime=Runtime(imdb)
```

```python
def isLeaf(tree):
    if tree[-1]==None:
        return True
    else:
        return False
```

```python
def printTree(tree):
    for i in tree:
        if isLeaf(i):
            print(f"|-{i[0]}--{i[1]}")
        else:
             print(f"-{i}")
```

```python
def inputprompt():
    while True:
        root_input=input("Which dataset do you want to use, imdb or golden globe award?")
        if root_input in ['IMDB','imdb','golden globe award','Golden Globe Award','GOLDEN GLOBE AWARD']:
            return root_input.lower()
            break
        else:
            print("Please input a correct answer")
            continue
```

```python
class imdb_Leaf:
    def __init__(self,tree,value=None):
        self.tree=tree
        self.value=value
    def leaf1(self):
        while True:
            print(f"The Genre contains Drama, Crime, Action, Adventure, Biography, History, Romance, Western, Thriller, Family, Animation, Comedy, Mystery, Music, War, Horror.")
            self.value=input(f"Which {self.tree[1][0]} do you want to watch?(hint:1 anwser)")
            if self.value in Genre:
                  return self.value
                  break
            else:
                print("Please input the answer from above")
                continue
    def leaf2(self):
            while True:
                print(f"The Runtime contains <=60, 60-120, 120-180, >180 (min).")
                self.value=input(f"Which {self.tree[2][0]}(min) do you want to watch?(hint:1 anwser)")
                if self.value in imdb_runtime:
                      return self.value
                      break
                else:
                    print("Please input the answer from above")
                    continue

    def leaf3(self):
        while True:
                print(f"The Release year contains 2010-2020, 2000-2010, 1990-2000, 1970-1990, before 1970.")
                self.value=input(f"Which {self.tree[3][0]} do you want to watch?(hint:1 anwser)")
                if self.value in release_year:
                      return self.value
                      break
                else:
                    print("Please input the answer from above")
                    continue
```

```python
class golden_globe:
    def __init__(self,tree,value=None):
        self.tree=tree
        self.value=value
    def leaf1(self):
        while True:
            print(f"The award category contains movie, television.")
            self.value=input(f"Which {self.tree[5][0]} do you want to watch?(hint:1 anwser)")
            if self.value in award_category:
                  return self.value
                  break
            else:
                print("Please input the answer from above")
                continue
    def leaf2(self):
            while True:
                print(f"The year is from 1944 to 2020.")
                self.value=input(f"Which {self.tree[6][0]} do you want to watch?(hint:1 anwser)")
                
                try:
                    if 1944<=int(self.value)<=2020:
                          return self.value
                          break
                    else:
                        print("Please input the answer from above")
                        continue
                except:
                    print("Please input the answer from above")
                    continue
    
```

```python
def Filter_data(tree):
    Recommendations=[]
    node=inputprompt()
    if node =="imdb":
        subtree=imdb_Leaf(tree)
        Node1=subtree.leaf1()
        genre=[]
        tree[1][1]=Node1
        for i in range(len(imdb["Genre"])):
            if Node1[0] in imdb["Genre"][i]:
                       genre.append(imdb["Series_Title"][i])
        tree[1][2]=genre
        Node2=subtree.leaf2()
        tree[2][1]=Node2
        leaf_runtime[Node2]
        tree[2][2]=leaf_runtime[Node2]
        Node3=subtree.leaf3()
        tree[3][1]=Node3
        leaf_release[Node3]
        tree[3][2]=leaf_release[Node3]
        for i in imdb["Series_Title"]:
            if i in genre and i in leaf_runtime[Node2]  and i in leaf_release[Node3]:
                    Recommendations.append(i)
        
        print(f"This is your search tree \n {printTree(tree)}")
        if len(Recommendations)>=5:
            Recommendations=Recommendations[:5]
        else:
            Recomendations=Recommendations
        if len(Recommendations)>0:
            print("Following is the recommendation for you:")
            for i in range(len(imdb["Series_Title"])):
                for rec in Recommendations:
                    if rec==imdb["Series_Title"][i]:
                        print(imdb.loc[i,:])
        else:print("No result,you can try again")
        return Recommendations
                       
    elif node == "golden globe award":
        subtree=golden_globe(tree)
        Node1=subtree.leaf1()
        tree[5][1]=Node1
        leaf_category[Node1]
        tree[5][2]=leaf_category[Node1]
        Node2=subtree.leaf2()
        film_year=[]
        tree[6][1]=Node2
        tree[6][2]=film_year
        for i in range(len(golden_data["year_award"])):
            if int(Node2)==golden_data["year_award"][i]:
                film_year.append(golden_data["film"][i])
        for i in golden_data["film"]:
            if i in leaf_category[Node1] and i in film_year:
                if i not in Recommendations:
                    Recommendations.append(i)
        for i in Recommendations:
            if i=='None':
                Recommendations.remove(i)
        print(f"This is your search tree \n {printTree(tree)}")
        if len(Recommendations)>=5:
            Recommendations=Recommendations[:5]
        else:
            Recomendations=Recommendations
        if len(Recommendations)>0:
            print("Following is the recommendation for you:")
            for i in range(len(golden_data["film"])):
                for rec in Recommendations:
                    if rec==golden_data["film"][i]:
                        print(golden_data.loc[i,["film","year_film","year_award","category","nominee"]])
        else:print("No result,you can try again")
        
        return Recommendations    
```
